#### Gathering the data from the .csv files

```python
import pandas as pd
import glob

# Getting file paths
files = sorted(glob.glob("Dates/All Holidays/Holidays - *.csv"))

# Starting year and empty df dict
year = 2016
dfs = {}

# Reading the files and saving them to a dict
for file in files:
    dfs[year] = pd.read_csv(file)
    year += 1
```

#### Extracting the summer holiday dates for each state

```python
summer_dfs = {}

for year,df in dfs.items():
    temp = df[['Bundesland', 'Sommerferien']].copy()
    # Remove asterisks
    temp['Sommerferien'] = temp['Sommerferien'].str.replace('*', '', regex=False)
    # split start and end
    temp[['start', 'end']] = temp['Sommerferien'].str.split(' - ', expand=True)
    
    # convert to datetime
    temp['start'] = pd.to_datetime(temp['start'] + str(year), format='%d.%m.%Y')
    temp['end'] = pd.to_datetime(temp['end'] + str(year), format='%d.%m.%Y')

    summer_dfs[year] = temp[['Bundesland', 'start', 'end']]
```

#### Saving the dates for each year

```python
for year, df in summer_dfs.items():
    df.to_csv(f"Dates/Summer/summer_holidays_{year}.csv", index = False)
```
