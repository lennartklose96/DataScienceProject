# PM10

### Get all stations located in Poland that measure PM10
Stations are sorted descending by the duration of their measurements

```python
api_url = "https://api.openaq.org/v3/locations"

params = {
    "iso": "AT",   # only consider stations located in Austria
    "parameters_id": 1,   # PM10 
    "monitor": True,   # only consider stations which only use devices that are used for official/governmental monitoring
    "limit": 1000        # results per page
}

headers = {
    "X-API-Key": "2e4741b1051f0a4545639472115170bf828e6a8569951f0969bbb98bbdb3863d"
}

response = requests.get(api_url, params=params, headers=headers)

if response.status_code == 200:
    data = response.json()

results = data["results"]

df = pd.DataFrame(results)

df["datetimeFirst"] = pd.to_datetime(df["datetimeFirst"].apply(lambda x: x["utc"]))
df["datetimeLast"] = pd.to_datetime(df["datetimeLast"].apply(lambda x: x["utc"]))

df_sorted = df.sort_values(
    by=["datetimeFirst", "datetimeLast"],
    ascending=[True, False]
)

df_top_pm10 = df_sorted

df_top_pm10
```

### Get sensor_ids from the stations
The daily measurements from a station can only be queried by the sensor_id.
A station usually only has one sensor for one parameter.

```python
rows = []

for _, station in df_top_pm10.iterrows():
    
    for sensor in station["sensors"]:
        
        # Nur PM10 behalten
        if sensor["parameter"]["id"] == 1:
            
            rows.append({
                "station_id": station["id"],
                "station_name": station["name"],
                "sensor_id": sensor["id"],
                "sensor_name": sensor["name"]
            })

df_pm10_sensors = pd.DataFrame(rows)

df_pm10_sensors
```

```python
sensor_ids = df_pm10_sensors["sensor_id"].tolist()
print(len(sensor_ids))
sensor_ids
```

### Get daily, average values from each sensor/station for 2016-2026

```python
headers = {
    "X-API-Key": "2e4741b1051f0a4545639472115170bf828e6a8569951f0969bbb98bbdb3863d"
}

all_daily_values = []

for sensor_id in sensor_ids:

    api_url = f'https://api.openaq.org/v3/sensors/{sensor_id}/measurements/daily'

    page = 1
    
    while True:
        params = {
            'datetime_to': '2026-03-01',
            'datetime_from': '2016-01-01',
            'limit': 1000,
            'page': page
        }
    
        response = requests.get(api_url, params=params, headers=headers)
    
        if response.status_code != 200:
            break
            
        data = response.json()
        results = data['results']
    
        if not results:
            break

        print(f"Sensor {sensor_id}: Seite {page}, {len(results)} Ergebnisse")
    
        for r in results:
            #sensor_results.append(r)

        # Datum (nur YYYY-MM-DD)
            date = r["period"]["datetimeFrom"]["local"][:10]

            # Daily Average
            value = r["summary"]["avg"]

            if value is not None:
                all_daily_values.append({
                    "date": date,
                    "value": value
                })
        
        page += 1

        time.sleep(0.2)

print(len(all_daily_values))
all_daily_values
```

### Transform queried data into a dataframe
Data is grouped by date and values are aggregated by mean

```python
df_results = pd.DataFrame(all_daily_values)

df_results['date'] = pd.to_datetime(df_results['date'])

france_daily = (
    df_results.groupby("date")
              .agg(france_pm10_mean=("value", "mean"))
              .reset_index()
              .sort_values("date")
)

france_daily
```

### Convert dataframe to csv

```python
france_daily[["date", "france_pm10_mean"]].to_csv(
    "daily_avg_austria_PM10.csv",
    index=False
)
```

## Cleaning

```python
df = pd.read_csv("../romania/2daily_avg_romania_PM10.csv")

df = df[
    (df['value'] >= 0) & (df['value'] <= 100) & 
    (
        (~df['date'].between('2019-07-05', '2019-12-23')) | 
        (df['value'] >= 2)
    )
]

df.to_csv('cleaned_daily_avg_romania_PM10.csv', index=False)
```

## Monthly Data

```python
df = pd.read_csv('../romania/cleaned_daily_avg_romania_PM10.csv')

df['date'] = pd.to_datetime(df['date'])

df['date'] = df['date'].dt.to_period('M')

df = df.groupby('date').agg(value=("value", "mean")).reset_index()

df.to_csv("monthly_avg_romania_PM10.csv", index=False)

```
