```python
import pandas as pd 
import plotly.express as px 
import dash_ag_grid as dag 
from dash import Dash, html, dcc, callback, Output, Input
import plotly.graph_objects as go
```

```python
# must rename the date column from 'date start' to 'date'
df_de = pd.read_csv('../../Collected Data/Question 1/daily_avg_deutschland_PM10.csv').rename(columns={'date start': 'date'})

df_aut = pd.read_csv('../../Collected Data/Question 8/austria/cleaned_daily_avg_austria_PM10.csv')
df_fr = pd.read_csv('../../Collected Data/Question 8/france/cleaned_daily_avg_france_PM10.csv')
df_it = pd.read_csv('../../Collected Data/Question 8/italy/cleaned_daily_avg_italy_PM10.csv')
df_pl = pd.read_csv('../../Collected Data/Question 8/poland/cleaned_daily_avg_poland_PM10.csv')
df_ro = pd.read_csv('../../Collected Data/Question 8/romania/cleaned_daily_avg_romania_PM10.csv')
df_es = pd.read_csv('../../Collected Data/Question 8/spain/cleaned_daily_avg_spain_PM10.csv')
df_gb = pd.read_csv('../../Collected Data/Question 8/uk/cleaned_daily_avg_uk_PM10.csv')

data = {
    'Germany': df_de,
    'Austria': df_aut,
    'France': df_fr,
    'Italy': df_it,
    'Poland': df_pl,
    'Romania': df_ro,
    'Spain': df_es,
    'UK': df_gb
}

for df in data.values():
    df["date"] = pd.to_datetime(df["date"])
    df["year"] = df["date"].dt.year

years = sorted(data["Germany"]["year"].unique())

app = Dash()

app.layout = html.Div([
    dcc.Graph(id='graph-with-slider1'),

    dcc.Slider(
        min(years),
        max(years),
        step=None,
        value=min(years),
        marks={str(year): str(year) for year in years},
        id='year-slider1'
    )
])


@callback(
    Output('graph-with-slider1', 'figure'),
    Input('year-slider1', 'value')
)

def update_figure(selected_year):

    fig = go.Figure()

    for state, df in data.items():

        filtered_df = df[df["year"] == selected_year]

        fig.add_trace(
            go.Scatter(
                x=filtered_df["date"],
                y=filtered_df["value"],
                mode="lines",
                name=state
            )
        )

    fig.update_layout(
        title="PM10 Air Pollution by Countries",
        xaxis_title="Date",
        yaxis_title="PM<sub>10</sub> µg/m³",
        width=1000,
        height=500
    )

    return fig


if __name__ == '__main__':
    app.run(debug=True)
```
