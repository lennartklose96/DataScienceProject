## This file builds aggregate measurements for daily, monthly and yearly stats of the PM10, PM2.5 values and the temperature/precipitation for the top 100 german cities

#### Import and loading the file

```python
import pandas as pd

# Reading measurements and converting data type to datetime
df = pd.read_csv("CollectedData/AllCitiesMeasuredDaily.csv")
df["date"] = pd.to_datetime(df["date"])
df
```

#### Daily stats

```python
# Giving us a daily data frame, dropping irrelevant columns
daily = df.drop(["population", "lat", "lon"], axis=1)

# Building the daily average across all cities adjusted for population
weighted_daily = (
    df.groupby("date").apply(
        lambda g: pd.Series({
            "pm10": (g["pm10"] * g["population"]).sum() / g["population"].sum(),
            "pm25": (g["pm25"] * g["population"]).sum() / g["population"].sum(),
            "peak_temperature": (g["peak_temperature"] * g["population"]).sum() / g["population"].sum(),
            "precipitation_mm": (g["precipitation_mm"] * g["population"]).sum() / g["population"].sum()
        })
    )
    .reset_index()
)

# Saving the data
daily.to_csv("CollectedData/Aggregates/DailyCity.csv", index = False)
weighted_daily.to_csv("CollectedData/Aggregates/DailyWeighted.csv", index = False)
```

#### Monthly Stats

```python
# Building monthly aggregates
df_m = df.copy()
df_m["date"] = df_m["date"].dt.to_period("M")

# Group by city and month
df_m = df_m.groupby(["city", "population", "date"]).agg({
    "pm10": "mean",
    "pm25": "mean",
    "peak_temperature": "mean",
    "precipitation_mm": "mean"
}).reset_index()

# Giving us a monhtly data frame, dropping irrelevant columns
monthly = df_m.drop(["population"], axis=1)

# Building the monthly average across all cities adjusted for population
weighted_monthly = (
    df_m.groupby("date").apply(
        lambda g: pd.Series({
            "pm10": (g["pm10"] * g["population"]).sum() / g["population"].sum(),
            "pm25": (g["pm25"] * g["population"]).sum() / g["population"].sum(),
            "peak_temperature": (g["peak_temperature"] * g["population"]).sum() / g["population"].sum(),
            "precipitation_mm": (g["precipitation_mm"] * g["population"]).sum() / g["population"].sum()
        })
    )
    .reset_index()
)

# Saving the data
monthly.to_csv("CollectedData/Aggregates/MonthlyCity.csv", index = False)
weighted_monthly.to_csv("CollectedData/Aggregates/MonthlyWeighted.csv", index = False)
```

#### Yearly stats

```python
# Building yearly aggregates
df_y = df.copy()
df_y["date"] = df_y["date"].dt.to_period("Y")

# Group by city and year
df_y = df_y.groupby(["city", "population", "date"]).agg({
    "pm10": "mean",
    "pm25": "mean",
    "peak_temperature": "mean",
    "precipitation_mm": "mean"
}).reset_index()

# Giving us a yearly data frame, dropping irrelevant columns
yearly = df_y.drop(["population"], axis=1)

# Building the yearly average across all cities adjusted for population
weighted_yearly = (
    df_y.groupby("date").apply(
        lambda g: pd.Series({
            "pm10": (g["pm10"] * g["population"]).sum() / g["population"].sum(),
            "pm25": (g["pm25"] * g["population"]).sum() / g["population"].sum(),
            "peak_temperature": (g["peak_temperature"] * g["population"]).sum() / g["population"].sum(),
            "precipitation_mm": (g["precipitation_mm"] * g["population"]).sum() / g["population"].sum()
        })
    )
    .reset_index()
)

# Saving the data
yearly.to_csv("CollectedData/Aggregates/YearlyCity.csv", index = False)
weighted_yearly.to_csv("CollectedData/Aggregates/YearlyWeighted.csv", index = False)
```
