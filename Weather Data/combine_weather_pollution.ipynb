#### Loading both of the relevant .csv files

```python
import pandas as pd

df_p = pd.read_csv("Measurements/Top100_DailyCityPollution.csv")
df_w = pd.read_csv("Measurements/Top100_DailyCityTempRain.csv")

# Merge df
df_combined = pd.merge(
    df_p,
    df_w,
    on = ["city", "date"],
    how = "inner"
)

# Reordering the attributes
df_combined = df_combined[["city", "population", "lat", "lon", "date", "pm10", "pm25", "peak_temperature", "precipitation_mm"]]
```

#### Save the data

```python
# Save the combined data source for all cities
df_combined.to_csv("AllCitiesMeasuredDaily.csv", index = False)

# Saving invidiual city data
for city, city_df in df_combined.groupby("city"):
    city_df.to_csv("CollectedData/Cities/" + city + "_DailyWeather.csv", index = False)
```
