#### Cleaning up the city names

```python
import pandas as pd
import glob
import os

# Getting the top 100 cities
df_cities = pd.read_csv("../top_100_cities.csv")
cities = df_cities["city"].tolist()

# Reading the (unclean) weather data for all cities
df_weather = pd.read_csv("TempRain/Weather_Cities_all.csv")

# Replaces "ae", "oe" and "ue" with "ä", "ö", "ü" respectively
def replace_umlauts(text):
    text = text.replace("ae", "ä")
    text = text.replace("oe", "ö")
    text = text.replace("ue", "ü")
    return text


# Mapping english city names to german ones
city_map = {
    "Cologne" : "Köln",
    "Munich": "München"
}
# Changing the names by mapping appropriate renaming functions
df_weather["city"] = df_weather["city"].replace(city_map)
df_weather["city"] = df_weather["city"].apply(replace_umlauts)

# Dropping cities that had no pollution data vailable
df_weather = df_weather[df_weather["city"].isin(cities)]
# Saving the temperature and rain data
df_weather.to_csv("Top100_DailyCityTempRain.csv", index = False)
```
