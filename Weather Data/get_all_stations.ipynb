```python
import requests
import pandas as pd

# 1) Fetch all stations
url = "https://luftdaten.umweltbundesamt.de/api/air-data/v4/stations/json"
response = requests.get(url)
data = response.json()
# Assuming data['data'] is a dict of station_id -> station_info_dict
stations_dict = data['data']
df_stations = pd.DataFrame.from_dict(stations_dict, orient='index')

# Rename columns
columns_order = [
    "id", "code", "name", "city", "synonym",
    "first_activity", "last_activity",
    "lon", "lat",
    "network_id", "station_setting_id", "station_type_id",
    "network_code", "network_name", "station_setting_name",
    "station_setting_short", "station_type_name",
    "street", "street_number", "zip_code"
]

# Make sure the number of columns matches the data
df_stations.columns = columns_order
df_stations.to_csv("stations.csv", index=False)
df_stations
```
