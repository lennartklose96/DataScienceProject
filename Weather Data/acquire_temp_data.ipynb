#### Imports and Getting the top 100 cities

```python
import requests
import pandas as pd
from time import sleep

# Reading location data about german cities
df = pd.read_csv(
    "DE.txt",
    sep='\t',
    header=None,
    names=[
        "geonameid", "name", "ascii", "lat", "lon", "feature_class", "feature_code",
        "country", "cc2", "admin1", "admin2", "admin3", "admin4",
        "population", "elevation", "dem", "timezone", "modification_date"
    ]
)
###########################
### CLEANING DATA FRAME ###
###########################

# Keep only populated places (feature_code starting with 'PPL')
df_cities = df[df["feature_code"].str.startswith("PPL")]
# Select relevant columns
df_cities = df_cities[["name", "lat", "lon", "population", "admin1"]]
# Take the top 100 most populated cities, divided into segments
df_top_cities = df_cities.sort_values("population", ascending=False)
df_cities_1 = df_top_cities.iloc[0:20]
df_cities_2 = df_top_cities.iloc[20:40]
df_cities_3 = df_top_cities.iloc[40:60]
df_cities_4 = df_top_cities.iloc[60:80]
df_cities_5 = df_top_cities.iloc[80:100]
```

#### Function to get weather data from certain cities

```python
# Returns the weather data the cities specified in the dataframe
def get_weather_data(cities):
    # Collection of weather data
    weather_data = []

    # Loop through each city
    for idx, row in cities.iterrows():
        city_name = row['name']
        lat = row['lat']
        lon = row['lon']
        
        # Get the peak temperature for each city from 2016-2025
        try:
            weather = requests.get(
                "https://archive-api.open-meteo.com/v1/archive",
                params = {
                    "latitude": lat,
                    "longitude": lon,
                    "start_date": "2016-01-01",
                    "end_date": "2025-12-31",
                    "daily": "temperature_2m_max",
                    "timezone": "Europe/Berlin"
                }
            ).json()
            
            # Convert to DataFrame
            city_df = pd.DataFrame({
                "city": city_name,
                "lat": weather['latitude'],
                "lon": weather['longitude'],
                "date": weather["daily"]["time"],
                "peak_temperature": weather["daily"]["temperature_2m_max"]
            })
            
            weather_data.append(city_df)
            # Help avoid rate limits
            sleep(1)
            print(f"Fetched weather for {city_name}")
        
        except Exception as e:
            print(f"Failed for {city_name}: {e}")
    return weather_data
```

Cities 1-20

```python
weather_data_1 = get_weather_data(df_cities_1)
```

Cities 21-40

```python
weather_data_2 = get_weather_data(df_cities_2)
```

Cities 41-60

```python
weather_data_3 = get_weather_data(df_cities_3)
```

Cities 61-80

```python
weather_data_4 = get_weather_data(df_cities_4)
```

Cities 81-100

```python
weather_data_5 = get_weather_data(df_cities_5)
```

#### Combine all the data into .csv

```python
# Combine all city dataframes
combined_weather_data = weather_data_1 + weather_data_2 + weather_data_3 + weather_data_4 + weather_data_5
df_weather_all = pd.concat(combined_weather_data, ignore_index=True)
df_weather_all.to_csv('PeakTempsTop100.csv', index = False)
```
