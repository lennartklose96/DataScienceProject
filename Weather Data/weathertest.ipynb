```python
import requests
import pandas as pd

r = requests.get("https://geocoding-api.open-meteo.com/v1/search?name=Berlin&countryCode=DE")
data = r.json()
lat = data['results'][0]['latitude']
lon = data['results'][0]['longitude']
print(lat)
print(lon)
weather = requests.get(
    "https://archive-api.open-meteo.com/v1/archive?",
    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": "2020-01-01",
        "end_date": "2024-12-31",
        "daily": "temperature_2m_max",
        "timezone": "Europe/Berlin"
    }
).json()

df = pd.DataFrame({
    "date": weather["daily"]["time"],
    "peak_temperature": weather["daily"]["temperature_2m_max"]
})
df
```
