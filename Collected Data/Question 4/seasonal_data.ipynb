#### Gathering daily data

```python
import pandas as pd

# Loading dataframes
df_pm10 = pd.read_csv("../Question 1/daily_avg_deutschland_PM10.csv")
df_pm25 = pd.read_csv("../Question 1/daily_avg_deutschland_PM2,5.csv")

# Function to clean date column and rename value
def clean_date_and_value(df, new_value_name):
    df = df.rename(columns={"date start": "date", "value": new_value_name})
    df["date"] = pd.to_datetime(df["date"]).dt.date  # keep only the date
    return df

# Apply cleaning and renaming
df_pm10 = clean_date_and_value(df_pm10, "pm10")
df_pm25 = clean_date_and_value(df_pm25, "pm25")

# Merge on date
df = pd.merge(df_pm10, df_pm25, on="date", how="outer")
# Convert date column to datetime if not already
df["date"] = pd.to_datetime(df["date"])

df
```

#### Splitting df up into seasons and saving them

```python
# Function to map month to season
def month_to_season(month):
    if month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    elif month in [9, 10, 11]:
        return "Autumn"
    else:  # 12, 1, 2
        return "Winter"

# Add a season column
df["season"] = df["date"].dt.month.apply(month_to_season)

# Create separate DataFrames per season
df_spring = df[df["season"] == "Spring"].copy()
df_summer = df[df["season"] == "Summer"].copy()
df_autumn = df[df["season"] == "Autumn"].copy()
df_winter = df[df["season"] == "Winter"].copy()

# Drop 'season' column
df_spring = df_spring.drop(columns="season")
df_summer = df_summer.drop(columns="season")
df_autumn = df_autumn.drop(columns="season")
df_winter = df_winter.drop(columns="season")

# Saving data
df_spring.to_csv("spring_daily_data.csv")
df_summer.to_csv("summer_daily_data.csv")
df_autumn.to_csv("autumn_daily_data.csv")
df_winter.to_csv("winter_daily_data.csv")
```
