```python
import requests
import urllib.parse as up
import pandas as pd
from pandas import json_normalize
import json
import glob
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from sklearn.linear_model import LinearRegression
from dash import Dash, html, dcc, callback, Output, Input
import plotly.express as px
import dash_ag_grid as dag
import plotly.graph_objects as go
```

```python
df_d_PM10 = pd.read_csv("corona_PM10.csv")
df_d_PM25 = pd.read_csv("corona_PM2,5.csv")
df_d_NO2 = pd.read_csv("corona_NO2.csv")
df_m_PM10 = pd.read_csv("monthly_avg_corona_PM10.csv")
df_m_PM25 = pd.read_csv("monthly_avg_corona_PM2,5.csv")
df_m_NO2 = pd.read_csv("monthly_avg_corona_NO2.csv")


data = {
    "daily avg PM<sub>10</sub>  ": df_d_PM10,
    "daily avg PM<sub>2.5</sub>": df_d_PM25,
    "daily avg NO<sub>2</sub>": df_d_NO2,
    "monthly avg PM<sub>10</sub>  ": df_m_PM10,
    "monthly avg PM<sub>2.5</sub>": df_m_PM25,
    "monthly avg NO<sub>2</sub>": df_m_NO2
}

fig = go.Figure()

for name, df in data.items():
    fig.add_trace(
        go.Scatter(
            x=df["date start"],
            y=df["value"],
            mode="lines",
            name=name
        )
    )

fig.update_layout(
    title="Air Pollution Germany",
    xaxis_title="Date",
    yaxis_title="Concentration in µg/m³",

)

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Air Pollution Dashboard Germany"),
    dcc.Graph(figure=fig)
])

if __name__ == "__main__":
    app.run(debug=True)
```

```python
df_m_PM10 = pd.read_csv("monthly_avg_corona_PM10.csv")
df_m_PM25 = pd.read_csv("monthly_avg_corona_PM2,5.csv")
df_m_NO2 = pd.read_csv("monthly_avg_corona_NO2.csv")


data = {
    "monthly avg PM<sub>10</sub>  ": df_m_PM10,
    "monthly avg PM<sub>2.5</sub>": df_m_PM25,
    "monthly avg NO<sub>2</sub>": df_m_NO2
}


for df in data.values():
    df["date start"] = pd.to_datetime(df["date start"])

fig = go.Figure()

for name, df in data.items():

    fig.add_trace(
        go.Bar(
            x=df["date start"],
            y=df["value"],
            name=name,
        )
    )

    x_num = df["date start"].map(pd.Timestamp.toordinal)
    y = df["value"]

    coeff = np.polyfit(x_num, y, 3)
    poly = np.poly1d(coeff)

    y_reg = poly(x_num)

    fig.add_trace(
        go.Scatter(
            x=df["date start"],
            y=y_reg,
            mode="lines",
            name=name + " trend",
        )
    )

fig.update_layout(
    title="Air Pollution Germany",
    xaxis_title="Date",
    yaxis_title="Concentration in µg/m³",
)

app = Dash(__name__)

app.layout = html.Div([
    html.H1("Air Pollution Dashboard Germany"),
    dcc.Graph(figure=fig)
])

if __name__ == "__main__":
    app.run(debug=True)
```
