```python
import pandas as pd
import plotly.graph_objects as go

# 16 Bundesländer laden
files = [
    "PM10/daily_avg_baden-württemberg_PM10.csv",
    "PM10/daily_avg_bayern_PM10.csv",
    "PM10/daily_avg_berlin_PM10.csv",
    "PM10/daily_avg_brandenburg_PM10.csv",
    "PM10/daily_avg_bremen_PM10.csv",
    "PM10/daily_avg_hamburg_PM10.csv",
    "PM10/daily_avg_hessen_PM10.csv",
    "PM10/daily_avg_mecklenburg-vorpommern_PM10.csv",
    "PM10/daily_avg_niedersachsen_PM10.csv",
    "PM10/daily_avg_nordrhein-westfalen_PM10.csv",
    "PM10/daily_avg_rheinland-pfalz_PM10.csv",
    "PM10/daily_avg_saarland_PM10.csv",
    "PM10/daily_avg_sachsen_PM10.csv",
    "PM10/daily_avg_sachsen-anhalt_PM10.csv",
    "PM10/daily_avg_schleswig-holstein_PM10.csv",
    "PM10/daily_avg_thüringen_PM10.csv"
]

states = [
    "Baden-Württemberg", "Bayern", "Berlin", "Brandenburg", "Bremen", "Hamburg",
    "Hessen", "Mecklenburg-Vorpommern", "Niedersachsen", "Nordrhein-Westfalen",
    "Rheinland-Pfalz", "Saarland", "Sachsen", "Sachsen-Anhalt",
    "Schleswig-Holstein", "Thüringen"
]

# Daten einlesen und vorbereiten
data = {}
for state, file in zip(states, files):
    df = pd.read_csv(file)
    df["date start"] = pd.to_datetime(df["date start"])
    df["year"] = df["date start"].dt.year
    data[state] = df

# Alle Jahre sammeln
years = sorted(data[states[0]]["year"].unique())

# Figure vorbereiten
fig = go.Figure()

# Traces hinzufügen: einmal pro Bundesland pro Jahr
for year in years:
    for state in states:
        df_filtered = data[state][data[state]["year"] == year]
        fig.add_trace(go.Scatter(
            x=df_filtered["date start"],
            y=df_filtered["value"],
            name=state,
            visible=(year == years[0])  # nur erstes Jahr sichtbar
        ))

# Dropdown für Jahre bauen
buttons = []
num_states = len(states)
for i, year in enumerate(years):
    # Sichtbarkeit für alle Traces vorbereiten
    visibility = [False] * (len(years) * num_states)
    for j in range(num_states):
        visibility[i * num_states + j] = True
    buttons.append(dict(
        label=str(year),
        method="update",
        args=[{"visible": visibility},
              {"title": f"PM10 Air Pollution in {year}"}]
    ))

fig.update_layout(
    title=f"PM10 Air Pollution in {years[0]}",
    xaxis_title="Date",
    yaxis_title="PM10 µg/m³",
    updatemenus=[dict(active=0, buttons=buttons, x=0.1, y=1.15)]
)

# HTML exportieren
fig.write_html("pm10_germany_dropdown.html")
print("Fertige HTML-Datei: pm10_germany_dropdown.html")
```
